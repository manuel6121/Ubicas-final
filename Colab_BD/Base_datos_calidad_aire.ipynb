```python
#Se importan las librerias necesarias
import pandas as pd
import numpy as np

# Crea el df vacio
columnas = ['Temperatura', 'Humedad', 'CO', 'Alcohol', 'CO2', 'Tolueno', 'NH4', 'Acetona', 'CA']
Dataset_aire = pd.DataFrame(columns=columnas)

Dataset_aire
```

```python
#Se llena el dataset con datos aleatorios teniendo en cuenta valores estudiados para los quimicos y pruebas previamente realizadas
# Generar datos aleatorios que se quieren crear
num_datos = 1000

# Se generan datos aleatorios con porcentajes modificados para cada columna
#--------------------------Temperatura---------------------------------------
rango_temperatura_1 = np.random.uniform(14, 25, int(num_datos * 0.9))
rango_temperatura_2 = np.random.uniform(25, 35, int(num_datos * 0.1))
#Se juntan los valores
temperatura = np.concatenate((rango_temperatura_1, rango_temperatura_2))
#Se mezclan los valores
np.random.shuffle(temperatura)
#Se asignan a la clumna deseada
Dataset_aire['Temperatura'] = temperatura

#--------------------------Humedad---------------------------------------
rango_humedad_1 = np.random.uniform(14, 25, int(num_datos * 0.9))
rango_humedad_2 = np.random.uniform(25, 35, int(num_datos * 0.1))
#Se juntan los valores
humedad = np.concatenate((rango_humedad_1, rango_humedad_2))
#Se mezclan los valores
np.random.shuffle(humedad)
#Se asignan a la clumna deseada
Dataset_aire['Humedad'] = humedad

#--------------------------CO---------------------------------------
rango_CO_1 = np.random.uniform(3, 100, int(num_datos * 0.9))
rango_CO_2 = np.random.uniform(0, 10000, int(num_datos * 0.1))
#Se juntan los valores
CO = np.concatenate((rango_CO_1, rango_CO_2))
#Se mezclan los valores
np.random.shuffle(CO)
#Se asignan a la clumna deseada
Dataset_aire['CO'] = CO

#--------------------------Alcohol---------------------------------------
rango_Alcohol_1 = np.random.uniform(1, 20, int(num_datos * 0.9))
rango_Alcohol_2 = np.random.uniform(0, 1000, int(num_datos * 0.1))
#Se juntan los valores
Alcohol = np.concatenate((rango_Alcohol_1, rango_Alcohol_2))
#Se mezclan los valores
np.random.shuffle(Alcohol)
#Se asignan a la clumna deseada
Dataset_aire['Alcohol'] = Alcohol

#--------------------------CO2---------------------------------------
rango_CO2_1 = np.random.uniform(400, 450, int(num_datos * 0.9))
rango_CO2_2 = np.random.uniform(400, 2000, int(num_datos * 0.1))
#Se juntan los valores
CO2 = np.concatenate((rango_CO2_1, rango_CO2_2))
#Se mezclan los valores
np.random.shuffle(CO2)
#Se asignan a la clumna deseada
Dataset_aire['CO2'] = CO2

#--------------------------Tolueno---------------------------------------
rango_Tolueno_1 = np.random.uniform(0, 20, int(num_datos * 0.9))
rango_Tolueno_2 = np.random.uniform(0, 500, int(num_datos * 0.1))
#Se juntan los valores
Tolueno = np.concatenate((rango_Tolueno_1, rango_Tolueno_2))
#Se mezclan los valores
np.random.shuffle(Tolueno)
#Se asignan a la clumna deseada
Dataset_aire['Tolueno'] = Tolueno

#--------------------------NH4---------------------------------------
rango_NH4_1 = np.random.uniform(4, 60, int(num_datos * 0.9))
rango_NH4_2 = np.random.uniform(4, 500, int(num_datos * 0.1))
#Se juntan los valores
NH4 = np.concatenate((rango_NH4_1, rango_NH4_2))
#Se mezclan los valores
np.random.shuffle(NH4)
#Se asignan a la clumna deseada
Dataset_aire['NH4'] = NH4

#--------------------------Acetona---------------------------------------
rango_Acetona_1 = np.random.uniform(0, 20, int(num_datos * 0.9))
rango_Acetona_2 = np.random.uniform(0, 300, int(num_datos * 0.1))
#Se juntan los valores
Acetona = np.concatenate((rango_Acetona_1, rango_Acetona_2))
#Se mezclan los valores
np.random.shuffle(Acetona)
#Se asignan a la clumna deseada
Dataset_aire['Acetona'] = Acetona

Dataset_aire
```

```python
#Se definen los parametros para llenar la ultima columna CA se pondra en 0
#Cuando no haya ningun valor que afecte la calidad del aire y en 1 cuando lo haya

# Definir una función para asignar valores a la columna AC
def asignar_valor(row):
    if row['CO'] > 100 or row['Alcohol'] > 20 or row['CO2'] > 450 or row['Tolueno'] > 20 or row['NH4'] > 60 or row['Acetona'] > 20:
        return 1
    else:
        return 0

# Aplicar la función a cada fila del dataset para generar los valores de la columna AC
Dataset_aire['CA'] = Dataset_aire.apply(asignar_valor, axis=1)

Dataset_aire
```

```python
#OPCIONAL DAÑO DHT11
columnas_eliminar = ['Temperatura', 'Humedad']
ds_air = Dataset_aire.drop(columns=columnas_eliminar)
ds_air
```

```python
#Paso los archivos a .csv
nombre_archivo = 'ds_air.csv'
ds_air.to_csv(nombre_archivo, index=False)
```
